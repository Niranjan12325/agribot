```python
import pandas as pd

# Load dataset (change file type accordingly)
df = pd.read_excel("/content/agriculture_crop_data.xlsx")

# Normalize crop names once (not inside function)
df['Crop'] = df['Crop'].str.strip().str.lower()

# Chatbot function
def agri_bot(seed_input):
    seed_input = seed_input.strip().lower()

    if seed_input in df['Crop'].values:
        row = df[df['Crop'] == seed_input].iloc[0]

        response = f"""
🌱 Crop Details:

Crop: {row['Crop'].title()}
Best Weather: {row['Best_Weather']}
Time to Harvest: {row['Time_to_Harvest']}
Total Cost: {row['Total_Cost']}
People Required: {row['People_Required']}
Fertilizer: {row['Fertilizer']}
"""
        return response

    else:
        suggestions = ", ".join(df['Crop'].str.title().unique())
        return f"❌ Crop not found.\nTry one of these: {suggestions}"

# Run chatbot
print("👨‍🌾 Agriculture Chatbot Started (type 'exit' to stop)\n")

while True:
    seed = input("Enter crop name: ")

    if seed.lower() == "exit":
        print("👋 Chatbot session ended.")
        break

    print(agri_bot(seed))
```
